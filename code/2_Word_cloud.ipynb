```python
from wordcloud import WordCloud
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from bs4 import element

import spacy
nlp = spacy.load('en_core_web_sm')

from collections import Counter

import matplotlib.pyplot as plt
```

```python
df_in = pd.read_excel('../data/output/SS_Extracted_content.xlsx')[["content"]]
```

```python
all_contents = ""
for index, row in df_in.iterrows():
    row_content = row.content
    
    nlp_content = nlp( row_content)
    
    tokens = [token.text for token in nlp_content if not token.is_stop and not token.is_punct]
    content_str = " ".join( tokens)
    
    all_contents += " " + content_str
    
    if index % 100 == 0:
        print( index, ", ", end="")
    
print()
print( len(all_contents))
```

```python
wordcloud = WordCloud(max_font_size=40, background_color=(58,128,206), width=400, height=200).generate(all_contents)
wordcloud.to_file( "../data/output/0_wordcloud.bmp")
```

```python
df_in = pd.read_excel('../data/output/2_keywords_list.xlsx')[["keywords"]]
```

```python
keywords = ""
for index, row in df_in.iterrows():
    row_keywords = row.keywords
    
    keywords += ", " + row_keywords
    
    if index % 100 == 0:
        print( index, ", ", end="")
    
print()
print( len(keywords))
```

```python
wordcloud = WordCloud(max_font_size=40, background_color=(58,128,206), width=400, height=200).generate(keywords)
wordcloud.to_file( "../data/output/0_wordcloud_keywords.bmp")
```

```python
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
